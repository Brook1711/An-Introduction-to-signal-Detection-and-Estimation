```python
%matplotlib
import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np
```

```python
%matplotlib inline
# HW-1 2(a)
x = np.linspace(0, 1, 100)

fig = plt.figure()
plt.plot(x, 3/(2*(x +1)), '-', label = 'L(y)')
plt.plot(x, [1]*len(x), '-', label = 'tau')
plt.legend(['L(y)', 'tau'])
plt.xlabel('value of y')
plt.ylabel('L(y) and tau')
# plt.plot(x, np.cos(x), '--');
plt.ylim(0, 2)
```
