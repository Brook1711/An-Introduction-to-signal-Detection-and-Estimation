```python
%matplotlib
import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np
import math
```

```python
%matplotlib inline
# HW-1 2(a)
x = np.linspace(0, 1, 100)

fig = plt.figure()
plt.plot(x, 3/(2*(x +1)), '-', label = 'L(y)')
plt.plot(x, [1]*len(x), '-', label = 'tau')
plt.legend(['L(y)', 'tau'])
plt.xlabel('value of y')
plt.ylabel('L(y) and tau')
# plt.plot(x, np.cos(x), '--');
plt.ylim(0, 2)
```

```python
x = np.linspace(0, 1, 100)

fig = plt.figure()
plt.plot(x, (x/(1-x)), '-')
# plt.plot(x, [1]*len(x), '-', label = 'tau')
plt.legend(['\tau'])
plt.xlabel('value of \tau')
plt.ylabel('')
# plt.plot(x, np.cos(x), '--');
# plt.ylim(0, 2)
```

```python
%matplotlib inline
# HW-1 2(a)
x = np.linspace(0, 10, 100)

fig = plt.figure()
plt.plot(x, np.power(math.e, -x), '-', label = r'$p_0(y)$')
plt.plot(x, x*np.power(math.e, -x), '-', label = 'tau')
plt.plot(x, x, label = 'tau')
plt.legend([r'$p_0(y)$', r'$p_1(y)$', r'$L(y)$'])
plt.xlabel(r'value of $y$')
plt.ylabel('value')
# plt.plot(x, np.cos(x), '--');
plt.xlim(0, 10)
plt.ylim(0, 2)
```

```python
%matplotlib inline
# HW-1 2(a)
x = np.linspace(-10, 10, 100)

fig = plt.figure()
plt.plot(x, np.power(math.e, -np.absolute(x)), '-', label = r'$p_0(y)$')
plt.plot(x, np.power(math.e, -np.absolute(x-1)), '-', label = 'tau')
# plt.plot(x, x, label = 'tau')
plt.legend([r'$p_0(y)$', r'$p_1(y)$', r'$L(y)$'])
plt.xlabel(r'value of $y$')
plt.ylabel('value')
# plt.plot(x, np.cos(x), '--');
plt.xlim(-10, 10)
plt.ylim(0, 1.2)
```
